```python
import numpy as np
import cv2
# Will use matplotlib for showing the image
from matplotlib import pyplot as plt
# Plot inline
%matplotlib inline
```

### Creating a black image

```python
black_img = np.zeros([150, 200, 1], 'uint8')
```

```python
cv2.imshow("Black", black_img)
print(black_img[0,0,:])

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Plot
plt.imshow(black_img)
```

### Creating a white image

```python
white = np.ones([150, 200, 3], 'uint16')
white *= (2 ** 16 - 1)
white = cv2.cvtColor(white, cv2.COLOR_BGR2RGB)
plt.imshow(white);
```

### Reading an image from file

```python
img = cv2.imread('./images/Geneva_Emotion_Wheel_-_English.png', 1)
```

```python
plt.imshow(img);
```

```python
height, width, channels = img.shape
print(f"Height: {height}, Width:{width}, Channels:{channels}")
```

### breaking into 3 channels:

```python
r, g, b = cv2.split(img) # red, gree, blue
```

```python
rgb_split = np.empty([height, width*3, 3], 'uint8')
```

```python
rgb_split[:,0:width]         = cv2.merge([b,b,b]) # blue
rgb_split[:,width:width*2]   = cv2.merge([g,g,g]) # green
rgb_split[:,width*2:width*3] = cv2.merge([r,r,r]) # red
plt.imshow(rgb_split);
```

### Hue, Saturation, Value (luminousity)

```python
hsv       = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
h, s, v   = cv2.split(hsv)
hsv_split = np.concatenate((h, s, v), axis=1)
plt.imshow(hsv_split);
```

### To gray

```python
grey = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(grey);
```

### Making one channel transparent

```python
b = img[:,:,0]
g = img[:,:,1]
r = img[:,:,2]

rgba = cv2.merge((b,g,r,g))
plt.imshow(rgba);
```

```python
rgba = cv2.merge((b,g,r,b))
plt.imshow(rgba);
```

```python
rgba = cv2.merge((b,g,r,r))
plt.imshow(rgba);
```
