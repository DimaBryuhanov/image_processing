```python
import numpy as np
import cv2
# Will use matplotlib for showing the image
from matplotlib import pyplot as plt
# Plot inline
%matplotlib inline
```

```python
img = cv2.imread('./images/Geneva_Emotion_Wheel_-_English.png', 1)
plt.imshow(img);
```

### Gaussian Blur

```python
blur = cv2.GaussianBlur(img, (5, 55), 0) # x=5 , y=55
plt.imshow(blur);
```

```python
kernel = np.ones((5,5), 'uint8')
```

### Dilation

```python
dilated = cv2.dilate(img, kernel, iterations=1)
plt.imshow(dilated);
```

### Erosion

```python
eroded = cv2.erode(img, kernel, iterations=1)
plt.imshow(eroded);
```

```python
eroded = cv2.erode(img, kernel, iterations=3)
plt.imshow(eroded);
```

### Scale

```python
img_half = cv2.resize(img, (0,0), fx=0.5, fy=0.5)
plt.imshow(img_half);
```

```python
img_strech = cv2.resize(img, (1500,1500))
plt.imshow(img_strech);
```

```python
img_strech_near = cv2.resize(img, (1500,1500), interpolation=cv2.INTER_NEAREST)
plt.imshow(img_strech_near);
```

### Rotation

```python
M = cv2.getRotationMatrix2D((0,0), -30, 1) #Rotating around top left corner
rotated = cv2.warpAffine(img, M, (img.shape[0], img.shape[1]))
plt.imshow(rotated);
```

```python
height, width, channels = img.shape
M = cv2.getRotationMatrix2D((height/2,width/2), -30, 1) # Rotating around the center
rotated = cv2.warpAffine(img, M, (img.shape[0], img.shape[1]))
plt.imshow(rotated);
```
